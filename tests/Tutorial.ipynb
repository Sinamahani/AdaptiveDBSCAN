# importing packages

```python
from dadbscan.density import EQ_Density
from dadbscan.clustering import dbscan
```

<b> Initiate density class</b>

```python
cell_number = 65

density = EQ_Density(cell_number, data_folder = r"ExampleData", data_file = r"decl_cat.csv")
heat_matrix = density.calc_density()
density.plot_den()
smooth_heat_matrix = density.cell_smoother(apply_smooth=True)
```

```python
radius = 0.092#density.radius

density_file_names = [f"ExampleData/den_decl_cat__65_smooth.csv", f"ExampleData/den_decl_cat__65.csv"]

for density_file_name in density_file_names:
    print(f"Clustering {density_file_name}")
    clustering = dbscan(radius, density_file_name)
    final = clustering.clustering()
    clustering.plot_clusters(shape_file_address='World_Countries_Generalized.shp')
    final.to_csv(f"Results/R__{density_file_name.split('/')[1]}")
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
```
